```python
'''Q. 1] What is a random variable in probability theory?
Ans. A random variable is a numerical quantity whose value depends on the outcome of a random experiment. It provides a way to quantify outcomes 
from a random process, mapping each possible outcome to a real number.

Key Characteristics

 1. Deterministic Mapping: A random variable assigns a single numerical value to each outcome in the sample space of a probability experiment.
 2. Types of Random Variables:
    A. Discrete Random Variable: Takes on a finite or countable number of distinct values (e.g., number of heads in coin tosses).
    B. Continuous Random Variable: Takes on an uncountable range of values, typically from an interval (e.g., time, weight, temperature).'''
```

```python
'''Q.2] What are the types of random variables?
Ans. In probability theory, random variables are classified into two main types based on the nature of the values they can take: Discrete Random 
Variables and Continuous Random Variables.

1. Discrete Random Variable
A discrete random variable can take on a finite or countably infinite set of distinct values.

Key Characteristics:
  a. Values are often whole numbers.
  b. Associated with discrete probability distributions like the Binomial, Poisson, or Geometric distributions.
  c. Probability is defined using a Probability Mass Function (PMF):
      𝑃(𝑋 = 𝑥) = Probability that 𝑋 equals 𝑥.
  Examples:
      Number of heads in 5 coin tosses: 
        𝑋 = 0, 1, 2, 3, 4, 5
      Number of cars passing through a toll booth in an hour: 
        𝑋 = 0, 1, 2, ….

2. Continuous Random Variable
A continuous random variable can take on an uncountable number of values, usually within an interval of real numbers.

Key Characteristics:
Values can include fractions, decimals, or irrational numbers.
Associated with continuous probability distributions like the Normal, Exponential, or Gamma distributions.
Probability is defined using a Probability Density Function (PDF):
𝑃(𝑎 ≤ 𝑋 ≤ 𝑏) = ∫𝑎𝑏 𝑓𝑋(𝑥) 𝑑𝑥.

(The probability of 𝑋 taking on a single value is always 0, as 𝑃(𝑋 = 𝑥) = ∫𝑥𝑥 𝑓𝑋 (𝑥) 𝑑𝑥 = 0).
Examples:
    Height of people in centimeters: 𝑋 ∈ [50, 250].
    Time taken to complete a task: 𝑋 > 0.
    
3. Special Types of Random Variables

    Categorical Random Variable:
        Takes on categorical values like "red," "blue," or "green."
        Example: Gender (𝑋 = "Male", "Female").
    Ordinal Random Variable:
        Takes on ordered categorical values.
        Example: Education level (𝑋 = "High School", "Bachelor’s", "Master’s".'''
```

```python
'''Q.4]  What are probability distribution functions (PDF)?
Ans. In probability theory, a Probability Distribution Function (PDF) represents the probability density of a continuous random variable over a 
range of values. Unlike discrete random variables, where probabilities are assigned to specific values, continuous random variables deal with ranges 
of values, and the PDF provides a way to calculate probabilities over those ranges.

Key Characteristics of PDFs

Definition:
    A PDF, denoted as 𝑓𝑋(𝑥), describes the relative likelihood of a continuous random variable 𝑋 taking on a particular value 𝑥.
    𝑓𝑋(𝑥) is not the probability of 𝑋 = 𝑥, but rather the density at 𝑥. The actual probability is calculated over an interval.

Probability Over an Interval:
    The probability of 𝑋 falling within an interval [𝑎, 𝑏] is given by:
        𝑃(𝑎 ≤ 𝑋 ≤ 𝑏) = ∫𝑎𝑏 𝑓𝑋(𝑥) 𝑑𝑥

Total Area Under the Curve:
    The total area under the PDF curve is always equal to 1:
        ∫−∞ ∞ 𝑓𝑋(𝑥) 𝑑𝑥 = 1

Non-Negative:
    The PDF is always non-negative:
        𝑓𝑋(𝑥) ≥ 0 ∀𝑥'''

print("Here is how to plot the PDF of a normal distribution in python")

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mean = 0
std_dev = 1

# Generate values
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, mean, std_dev)

# Plot the PDF
plt.figure(figsize=(6, 5))
plt.plot(x, pdf, label="Normal Distribution", color='blue')
plt.title("PDF of a Normal Distribution")
plt.xlabel("X")
plt.ylabel("Density")
plt.legend()
plt.grid()
plt.show()
```

```python
'''Q.8] What is the binomial distribution, and how is it used in probability?
Ans. The binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent 
Bernoulli trials, each with the same probability of success.

Key Properties of the Binomial Distribution
1. Parameters
    A binomial distribution is defined by two parameters:
        𝑛: Number of trials.
        𝑝: Probability of success in a single trial.

2. Outcomes
    The random variable 𝑋 represents the number of successes in 𝑛 trials. Possible values of 𝑋 are: 𝑋 ∈ {0, 1, 2, …, 𝑛}

3. Probability Mass Function (PMF)
    The probability of observing 𝑘 successes out of 𝑛 trials is given by:

        𝑃(𝑋 = 𝑘) =(𝑛|𝑘) 𝑝(^𝑘)(1 − 𝑝)^𝑛−𝑘
 
        where:

            (𝑛|𝑘) = 𝑛!/(𝑘!(𝑛 − 𝑘)!) is the number of ways to choose 𝑘 successes from 𝑛 trials.

4. Mean (Expected Value)
    The mean represents the expected number of successes:

        𝐸[𝑋] = 𝑛⋅𝑝

5. Variance
    The variance quantifies the spread of the distribution:

        Var(𝑋) = 𝑛⋅𝑝⋅(1 − 𝑝)

6. Assumptions
The binomial distribution relies on these key assumptions:

    Fixed number of trials (𝑛): The number of trials is predetermined.
    Independent trials: The outcome of one trial does not affect others.
    Constant probability (𝑝): The probability of success remains the same for each trial.
    Two outcomes: Each trial results in success (1) or failure (0).

When to Use the Binomial Distribution
Use the binomial distribution when:
    The number of trials is fixed.
    Each trial is independent.
    The outcome of each trial is binary (success or failure).
    You want to model the total number of successes in those trials.'''

print("Sample code for implementation in python")
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 10  # Number of trials
p = 0.7  # Probability of success

# Generate probabilities for all possible outcomes
x = np.arange(0, n + 1)  # Possible values: 0 to n
pmf = binom.pmf(x, n, p)  # PMF values

# Plot the PMF
plt.bar(x, pmf, color='skyblue', edgecolor='k', alpha=0.7)
plt.title("Binomial Distribution (n=10, p=0.7)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()

# Probability of exactly 7 successes
prob_7 = binom.pmf(7, n, p)
print(f"Probability of exactly 7 successes: {prob_7:.4f}")
```

```python
'''Q.9] What is the Poisson distribution and where is it applied?
Ans. The Poisson distribution is a discrete probability distribution that models the number of events occurring in a fixed interval of time, space, 
or another domain, where these events happen:
  1. Independently of each other.
  2. At a constant average rate (mean rate of occurrence per interval).
  3. With no two events occurring at the same instant (no clustering).

Key Properties of the Poisson Distribution
1. Probability Mass Function (PMF):
The probability of observing 𝑘 events in a given interval is:

𝑃(𝑋 = 𝑘) = (𝜆^𝑘.𝑒^−𝜆)/𝑘!,         𝑘 = 0, 1, 2, …

where:

𝜆: The average number of events in the interval (mean rate).
𝑘: The actual number of events.
𝑒: Euler's number (≈2.718).

2. Cummulative DIstribution Function (CDF):
The Cumulative Distribution Function (CDF) of a Poisson random variable 
𝑋 gives the probability that 𝑋, the number of events, is less than or equal to a given value 𝑘. Mathematically, it is defined as:

                      𝑘
𝐹(𝑘; 𝜆) = 𝑃(𝑋 ≤ 𝑘) = ∑  (𝜆^𝑖.𝑒^−𝜆)/𝑖!
                     𝑖=0


Where:

𝜆: The average rate (mean) of events in a given interval.
𝑘: The value up to which the probability is calculated.
𝑒: Euler's number (≈2.718).

3. Mean and Variance:
For a Poisson random variable 𝑋:  𝐸[𝑋] = 𝜆,  Var(𝑋) = 𝜆
The mean and variance of the Poisson distribution are equal.

Applications of the Poisson Distribution
The Poisson distribution is applied in scenarios where events occur randomly and independently over time, space, or another dimension. Common use cases 
include:

  1. Traffic and Queue Management:
    Number of cars passing through a toll booth in an hour.
    Number of customers arriving at a service desk per minute.

  2. Call Centers:
    Number of calls received by a call center in a given time frame.

  3. Healthcare:
    Number of patients arriving at an emergency room in an hour.

  4. Natural Events:
    Number of earthquakes in a region over a year.
    Number of meteorites striking Earth annually.
  5. Manufacturing:
    Number of defects found in a fixed length of material.'''

print("Sample python code")
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lambda_rate = 4  # Average rate of events per interval
k_values = np.arange(0, 15)  # Possible values of k (number of events)

# PMF of the Poisson distribution
pmf_values = poisson.pmf(k_values, lambda_rate)

# Plot
plt.bar(k_values, pmf_values, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Poisson Distribution (λ = 4)")
plt.xlabel("Number of Events (k)")
plt.ylabel("Probability")
plt.show()

# Probability of exactly 6 calls
prob_6_calls = poisson.pmf(6, lambda_rate)
print(f"Probability of receiving exactly 6 calls: {prob_6_calls:.4f}")
```

```python
'''Q.10] What is a continuous uniform distribution?
Ans. The continuous uniform distribution is a probability distribution in which all outcomes in a specified range are equally likely. It is 
characterized by a constant probability density over the interval [𝑎, 𝑏].

Key Properties
1. Probability Density Function (PDF): For a continuous random variable 𝑋 uniformly distributed over [𝑎, 𝑏], the PDF is given by:
f(x)= { 1/(b-a), if a≤x≤b
      { 0,    Otherwise

a: Lower bound of the range.
𝑏: Upper bound of the range.
𝑓(𝑥) is constant across the interval [𝑎, 𝑏].

2. Cumulative Distribution Function (CDF): The CDF, 𝐹(𝑥), is:

             {0,        if 𝑥<𝑎
      𝐹(𝑥) = {𝑥−𝑎/𝑏−𝑎,  if 𝑎≤𝑥≤𝑏
             {1,        if 𝑥>𝑏

3. Mean:

    𝐸[𝑋] = 𝑎+𝑏/2
 
The mean is the midpoint of the interval. 

4. Variance:

    Var(𝑋) = ((𝑏−𝑎)^2)/12
 
The variance depends on the square of the interval's width.

Applications of Continuous Uniform Distribution

1. Simulation: Random sampling or generating random numbers over a fixed range.
2. Quality Control: Modeling measurements assumed to be evenly distributed over a tolerance range.
3. Simple Random Processes: Representing events equally likely to occur over time or space.'''

print("Sample python code")
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Parameters
a = 9  # Lower bound
b = 17  # Upper bound
x = np.linspace(a - 1, b + 1, 1000)  # Range of x values

# PDF and CDF
pdf = uniform.pdf(x, loc=a, scale=b - a)
cdf = uniform.cdf(x, loc=a, scale=b - a)

# Plot PDF
plt.figure(figsize=(10, 5))
plt.plot(x, pdf, label='PDF', color='blue')
plt.title("Continuous Uniform Distribution PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.axvline(11, color='red', linestyle='--', label='11 AM')
plt.axvline(14, color='green', linestyle='--', label='2 PM')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Plot CDF
plt.figure(figsize=(10, 5))
plt.plot(x, cdf, label='CDF', color='orange')
plt.title("Continuous Uniform Distribution CDF")
plt.xlabel("x")
plt.ylabel("Cumulative Probability")
plt.axvline(11, color='red', linestyle='--', label='11 AM')
plt.axvline(14, color='green', linestyle='--', label='2 PM')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
'''Q.11] What are the characteristics of a normal distribution?
Ans. 

1. Symmetry:
The normal distribution is symmetric about its mean (𝜇).
The left and right halves of the curve are mirror images of each other.

2. Bell-shaped Curve:
The probability density function (PDF) forms a bell-shaped curve.
The highest point is at the mean (𝜇).

3. Mean, Median, and Mode:
All three are equal and located at the center of the distribution.

4. Spread Determined by Standard Deviation (𝜎):
Larger 𝜎 values result in a wider and flatter curve.
Smaller 𝜎 values produce a narrower and taller curve.

5. Empirical Rule (68-95-99.7 Rule):
68% of data falls within 1 standard deviation (𝜇±𝜎).
95% of data falls within 2 standard deviations (𝜇±2𝜎).
99.7% of data falls within 3 standard deviations (𝜇±3𝜎).

6. Tails Extend to Infinity:
The curve approaches, but never touches, the horizontal axis on both sides.
This means there is a nonzero probability for extreme values, although it diminishes rapidly.

7. Area Under the Curve:
The total area under the curve is always equal to 1.
This property ensures that the distribution represents a valid probability distribution.'''
```

```python
'''Q.12] What is the standard normal distribution, and why is it important?
Ans. The standard normal distribution is a special case of the normal distribution where:
    The mean (𝜇) is 0.
    The standard deviation (𝜎) is 1.
The standard normal distribution is represented by the random variable 𝑍, which is often referred to as the z-score. 

Characteristics of the Standard Normal Distribution

1. Mean = 0, Standard Deviation = 1:
Centered at 0, the curve is symmetric about this point.

2. Bell-shaped Curve:
The curve resembles a standard bell shape, similar to any normal distribution.

3. Z-scores:
The 𝑍-value indicates how many standard deviations a data point is from the mean.

Importance of Standard Normal Distribution:

1. Foundation for Z-scores:

The standard normal distribution allows us to calculate 𝑍-scores, which measure the relative position of a data point within a distribution:
       𝑍 = 𝑋−𝜇/𝜎
Where:
𝑋: The value in the original dataset.
𝜇: Mean of the original dataset.
𝜎: Standard deviation of the original dataset.

2. Simplifies Probability Calculations:
    Probabilities for any normal distribution can be transformed into probabilities for the standard normal distribution by converting data points to 
    𝑍-scores. This is useful because:
        Standard normal distribution tables (Z-tables) provide pre-calculated cumulative probabilities.
        These probabilities can then be interpreted for any normal distribution.

3. Central Role in Hypothesis Testing:
    Many statistical tests (e.g., 𝑍-tests) rely on the standard normal distribution for determining critical values and 𝑝-values.

4. Applications in Confidence Intervals:
    Confidence intervals for population means and proportions are often based on the standard normal distribution.

5. Universal Applicability via Standardization:
    By converting data to 𝑍-scores, we can compare values from different normal distributions.'''

print("Sample code for implementation in python")

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the standard normal distribution
mean = 3
std_dev = 1
x = np.linspace(0, 6, 1000)
pdf = norm.pdf(x, mean, std_dev)

# Z-score range (e.g., Z < -1)
z_value = 2
highlight = x[x < z_value]

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label='Standard Normal Distribution', color='blue')
plt.fill_between(highlight, norm.pdf(highlight, mean, std_dev), color='orange', alpha=0.5, label='P(Z < 2)')
plt.axvline(z_value, color='red', linestyle='--', label=f'Z = {z_value}')
plt.title('Standard Normal Distribution and Z-scores')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
'''Q.13] What is the Central Limit Theorem (CLT), and why is it critical in statistics?
Ans. The Central Limit Theorem (CLT) states that, regardless of the population's original distribution, the sampling distribution of the sample mean 
will tend to follow a normal distribution as the sample size becomes large enough. This holds true under the following conditions:

    The samples are randomly selected and independent.
    The sample size 𝑛 is sufficiently large (𝑛>30 is often used as a rule of thumb, but smaller 𝑛 may suffice if the population distribution is 
    approximately normal).

Formal Statement of CLT
If you draw a sufficiently large number of independent random samples of size 𝑛 from any population with:
    A mean 𝜇, 
    A finite standard deviation 𝜎,

then the sampling distribution of the sample mean 𝑋:

    1. Will have a mean equal to the population mean (𝜇𝑋ˉ=𝜇).
    2. Will have a standard deviation (also called the standard error)
    3. Will approach a normal distribution as 𝑛→∞.

Criticallity of Central Limit Theorem in Statistics:

1. Foundation for Statistical Inference:
    The CLT allows statisticians to make inferences about a population from sample data. For example, confidence intervals and hypothesis tests for the 
    population mean are based on the assumption of a normal sampling distribution.

2. Applicability to Non-Normal Populations:
    Even if the population data is skewed or has an irregular distribution, the sampling distribution of the mean will still approximate normality for 
    large enough sample sizes.

3. Simplifies Analysis:
    With the CLT, we can use the properties of the normal distribution to analyze data, even if the underlying population distribution is unknown or 
    complex.

4. Enables Practical Approximations:
    Many statistical tests (e.g., 𝑡-tests, 𝑧-tests) rely on normality assumptions for their validity. The CLT ensures these assumptions are met for 
    large samples.

5. Real-World Applications:
    The CLT is fundamental in quality control, opinion polling, financial modeling, and other fields where sample data is used to make predictions about 
    entire populations.'''
```

```python
'''Q.14]  How does the Central Limit Theorem relate to the normal distribution?
Ans. The Central Limit Theorem (CLT) explains how the normal distribution arises naturally in many situations involving probability and statistics. 
The relationship can be summarized as follows:

Key Connections

1. Normality of the Sampling Distribution:
According to the CLT, if you take random samples of size 
𝑛 from a population with any distribution (normal, skewed, uniform, etc.), the distribution of the sample means will approximate a normal distribution 
as 𝑛 becomes large.
This holds true regardless of the original population distribution.

2. Foundation of the Normal Distribution's Ubiquity:
The CLT is the reason the normal distribution is so common in real-world data analysis. Many natural and social phenomena result from the aggregation 
of small, independent random effects. When these effects combine, the outcomes tend to follow a normal distribution due to the CLT.

3. Role of Sample Size (𝑛):
Small 𝑛: If the sample size is small, the sampling distribution may still reflect the shape of the original population.
Large 𝑛: As 𝑛 increases, the sampling distribution of the sample mean becomes increasingly normal, even if the population distribution is highly skewed 
or irregular.

4. Application to Statistical Methods:
Many statistical methods (e.g., confidence intervals, hypothesis tests) assume normality. The CLT justifies this assumption because the sample means 
or totals tend to be normally distributed for large 𝑛.'''

print("Here's Python code to demonstrate how the sampling distribution of the mean becomes normal, even if the population is non-normal")

import numpy as np
import matplotlib.pyplot as plt

# Generate a non-normal population (e.g., exponential distribution)
population = np.random.exponential(scale=2, size=100000)

# Sample sizes
sample_sizes = [5, 30, 100]
num_samples = 1000

plt.figure(figsize=(12, 8))

for i, n in enumerate(sample_sizes, 1):
    sample_means = [np.mean(np.random.choice(population, n, replace=True)) for _ in range(num_samples)]
    plt.subplot(1, 3, i)
    plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='blue', edgecolor='k')
    plt.title(f"Sample Size = {n}")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")

plt.tight_layout()
plt.suptitle("Central Limit Theorem and Normal Distribution", fontsize=16, y=1.02)
plt.show()
```

```python
'''Q.16]  How do you calculate a Z-score, and what does it represent?
Ans. A Z-score, also called a standard score, measures how many standard deviations a data point is from the mean of a dataset. It helps in 
understanding the position of a value relative to the rest of the data in a standard normal distribution.

The Z-score for a data point 𝑥 is calculated using the formula:

  𝑍 = (𝑥 − 𝜇)/𝜎

Where:
𝑥: The data point
𝜇: The mean of the dataset
𝜎: The standard deviation of the dataset

If you're working with sample data, the formula becomes:
  
  Z= (x− Xˉ)/s

Where:

𝑋ˉ: Sample mean
𝑠: Sample standard deviation

Significance of Z-score

Positive Z-Score:
Indicates the data point is above the mean.
Example: 
𝑍 = +2 means the data point is 2 standard deviations above the mean.

Negative Z-Score:
Indicates the data point is below the mean.
Example: 
𝑍 = −1.5 means the data point is 1.5 standard deviations below the mean.

Zero Z-Score:
Indicates the data point is exactly at the mean.

Magnitude of Z-Score:
Larger magnitudes (positive or negative) indicate data points farther away from the mean.

Uses of Z-Scores

 1. Standardization:
    Allows comparison of values from different datasets or scales by converting them into a common scale (standard normal distribution).

 2. Outlier Detection:
    Values with ∣𝑍∣>3 are typically considered outliers.

 3. Probability Calculation:
    Z-scores are used with the standard normal distribution table to determine probabilities for hypothesis testing.

 4. Normal Distribution:
    Helps identify how likely it is to observe a value within a given range in a normal distribution.
'''

print("Sample python code for calculating z-score")

import numpy as np

# Example dataset
data = [70, 85, 90, 60, 75, 80, 95]

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Calculate Z-scores
z_scores = [(x - mean) / std_dev for x in data]

print("Z-scores:", z_scores)

    
```

```python
'''Q.21] What are the assumptions for applying the Central Limit Theorem?
Ans. Assumptions of the Central Limit Theorem

  1. Independent Observations:
    The data points in the sample should be independent of one another.
    This means the selection of one observation should not influence the selection of another.

  2. Random Sampling:
    The sample should be randomly drawn from the population to avoid bias.
    Random sampling ensures the representativeness of the population.

  3. Finite Population Variance:
    The population from which the sample is drawn must have a finite variance (𝜎^2 < ∞).
    If the population variance is infinite, the CLT does not hold.

  4. Sample Size:
    The sample size should be sufficiently large for the CLT to apply.
      A common rule of thumb is 𝑛≥30, though this depends on the population's shape:
        If the population is symmetric or nearly normal, a smaller 𝑛 (e.g., 𝑛≥10) may suffice.
        If the population is highly skewed, a larger 𝑛 is needed for the sample mean to approximate normality.

  5. Identically Distributed Observations (for simple random sampling):
    The data points should come from the same distribution with the same mean and variance.'''
```

```python
'''Q.15] Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def verify_clt_normal_distribution(population_mean, population_std, sample_size, num_samples):
    """
    Simulates multiple samples from a normal distribution to verify the Central Limit Theorem (CLT).

    Parameters:
    - population_mean: Mean of the population normal distribution.
    - population_std: Standard deviation of the population normal distribution.
    - sample_size: Size of each sample.
    - num_samples: Number of samples to draw.

    Returns:
    - sample_means: List of means from all samples.
    """
    # Generate sample means
    sample_means = [
        np.mean(np.random.normal(population_mean, population_std, sample_size))
        for _ in range(num_samples)
    ]
    
    # Plot the population distribution
    x = np.linspace(population_mean - 4*population_std, population_mean + 4*population_std, 1000)
    y = norm.pdf(x, loc=population_mean, scale=population_std)
    
    plt.figure(figsize=(14, 6))

    plt.subplot(1, 2, 1)
    plt.plot(x, y, color='blue', label="Population Distribution")
    plt.fill_between(x, y, color='skyblue', alpha=0.6)
    plt.title("Population Normal Distribution")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()

    # Plot the sampling distribution of the mean
    plt.subplot(1, 2, 2)
    plt.hist(sample_means, bins=30, color='lightcoral', alpha=0.7, edgecolor='k', density=True)
    sampling_std = population_std / np.sqrt(sample_size)
    sampling_y = norm.pdf(x, loc=population_mean, scale=sampling_std)
    plt.plot(x, sampling_y, color='darkred', label="Sampling Distribution (Normal Approx.)")
    plt.title(f"Sampling Distribution of the Mean\n(Sample Size = {sample_size})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.legend()

    plt.tight_layout()
    plt.show()

    return sample_means

# Parameters
population_mean = 50       # Mean of the population
population_std = 15        # Standard deviation of the population
sample_size = 30           # Size of each sample
num_samples = 1000         # Number of samples

# Verify CLT
sample_means = verify_clt_normal_distribution(population_mean, population_std, sample_size, num_samples)
```

```python
'''Q.17] Generate random variables and calculate their corresponding probabilities using the binomial distribution'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def simulate_binomial_distribution(n, p, size):
    """
    Simulate random variables and calculate their probabilities using the binomial distribution.

    Parameters:
    - n: Number of trials
    - p: Probability of success in each trial
    - size: Number of random variables to generate
    """
    # Generate random variables from a binomial distribution
    random_variables = np.random.binomial(n, p, size)
    
    # Calculate unique values and their corresponding probabilities
    unique_values, counts = np.unique(random_variables, return_counts=True)
    probabilities = counts / size

    # Calculate the theoretical PMF
    x = np.arange(0, n + 1)  # Possible values (0 to n)
    theoretical_pmf = binom.pmf(x, n, p)

    # Plot the probabilities
    plt.figure(figsize=(12, 6))

    # Histogram of generated random variables
    plt.bar(unique_values, probabilities, color='skyblue', width=0.6, label="Empirical PMF", alpha=0.7)

    # Plot the theoretical PMF
    plt.plot(x, theoretical_pmf, 'o-', color='red', label="Theoretical PMF")

    # Add labels, title, and legend
    plt.title(f"Binomial Distribution (n={n}, p={p})")
    plt.xlabel("Number of Successes")
    plt.ylabel("Probability")
    plt.xticks(range(n + 1))
    plt.legend()
    plt.grid(alpha=0.3)

    # Show the plot
    plt.show()

    # Print results
    print("Unique values and their probabilities (Empirical PMF):")
    for value, prob in zip(unique_values, probabilities):
        print(f"Value: {value}, Probability: {prob:.4f}")
    
    print("\nTheoretical PMF values:")
    for value, prob in zip(x, theoretical_pmf):
        print(f"Value: {value}, Probability: {prob:.4f}")

# Parameters
n = 10  # Number of trials
p = 0.5  # Probability of success in each trial
size = 10000  # Number of random variables to generate

# Call the function
simulate_binomial_distribution(n, p, size)
```

```python
'''Q.18] Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, mean, std_dev):
    """
    Calculate the Z-score for a given data point.

    Parameters:
    - data_point: The value for which Z-score is to be calculated
    - mean: The mean of the dataset
    - std_dev: The standard deviation of the dataset

    Returns:
    - z_score: The calculated Z-score
    """
    z_score = (data_point - mean) / std_dev
    return z_score

def compare_to_standard_normal(z_score):
    """
    Compare a Z-score to a standard normal distribution.

    Parameters:
    - z_score: The Z-score to compare
    """
    # Plot the standard normal distribution
    x = np.linspace(-4, 4, 500)
    y = norm.pdf(x, 0, 1)  # Standard normal distribution (mean=0, std=1)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, color='blue', label='Standard Normal Distribution')

    # Mark the Z-score on the plot
    plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score: {z_score:.2f}')
    plt.fill_between(x, 0, y, where=(x <= z_score), color='red', alpha=0.3, label='Area under curve (<= Z-score)')

    # Add labels, legend, and title
    plt.title("Standard Normal Distribution and Z-score")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(alpha=0.3)

    plt.show()

    # Calculate cumulative probability
    cumulative_probability = norm.cdf(z_score)
    print(f"The Z-score is: {z_score:.2f}")
    print(f"The cumulative probability corresponding to Z-score {z_score:.2f} is: {cumulative_probability:.4f}")

# Example: Dataset and data point
mean = 50  # Mean of the dataset
std_dev = 10  # Standard deviation of the dataset
data_point = 65  # Data point to evaluate

# Calculate Z-score and compare to standard normal
z_score = calculate_z_score(data_point, mean, std_dev)
compare_to_standard_normal(z_score)
```

```python
'''Q.19] Implement hypothesis testing using Z-statistics for a sample dataset'''
import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05, alternative='two-sided'):
    """
    Perform hypothesis testing using Z-statistics.

    Parameters:
    - sample: List or array of sample data
    - population_mean: The mean of the population
    - population_std: The standard deviation of the population
    - alpha: Significance level (default: 0.05)
    - alternative: Type of test ('two-sided', 'less', 'greater')

    Returns:
    - Z-statistic
    - p-value
    - Decision: Reject or fail to reject the null hypothesis
    """
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Calculate Z-statistic
    standard_error = population_std / np.sqrt(sample_size)
    z_statistic = (sample_mean - population_mean) / standard_error

    # Calculate p-value
    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_statistic)))
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_statistic)
    elif alternative == 'less':
        p_value = norm.cdf(z_statistic)
    else:
        raise ValueError("alternative must be 'two-sided', 'greater', or 'less'")

    # Decision
    if p_value < alpha:
        decision = "Reject the null hypothesis"
    else:
        decision = "Fail to reject the null hypothesis"

    return z_statistic, p_value, decision


# Example Dataset
sample = [50, 52, 48, 49, 51, 47, 53, 50, 51, 49]  # Sample data
population_mean = 50  # Population mean
population_std = 2  # Population standard deviation
alpha = 0.05  # Significance level
alternative = 'two-sided'  # Test type

# Perform Z-test
z_statistic, p_value, decision = z_test(sample, population_mean, population_std, alpha, alternative)

# Print Results
print(f"Z-statistic: {z_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Decision: {decision}")
```

```python
'''Q.20] Create a confidence interval for a dataset using Python and interpret the result.'''
import numpy as np
from scipy.stats import norm

def calculate_confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for a given dataset.
    
    Parameters:
    - data: List or array of numerical data
    - confidence_level: Desired confidence level (default: 0.95 for 95%)
    
    Returns:
    - Tuple containing the lower and upper bounds of the confidence interval
    """
    # Calculate sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
    sample_size = len(data)

    # Calculate standard error
    standard_error = sample_std / np.sqrt(sample_size)

    # Find Z-score for the given confidence level
    z_score = norm.ppf((1 + confidence_level) / 2)

    # Calculate confidence interval
    margin_of_error = z_score * standard_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound


# Example Dataset
data = [50, 52, 48, 49, 51, 47, 53, 50, 51, 49]  # Sample data
confidence_level = 0.95  # 95% confidence level

# Calculate confidence interval
lower_bound, upper_bound = calculate_confidence_interval(data, confidence_level)

# Print Results
print(f"Sample Mean: {np.mean(data):.2f}")
print(f"Confidence Interval ({confidence_level * 100:.0f}%): [{lower_bound:.2f}, {upper_bound:.2f}]")
```

```python
'''Q.21] Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.'''
import numpy as np
from scipy.stats import norm

def confidence_interval_from_normal_data(mean, std_dev, size, confidence_level=0.95):
    """
    Generate data from a normal distribution and calculate the confidence interval for the mean.

    Parameters:
    - mean: Mean of the normal distribution
    - std_dev: Standard deviation of the normal distribution
    - size: Number of data points to generate
    - confidence_level: Desired confidence level (default: 0.95 for 95%)

    Returns:
    - lower_bound, upper_bound: Confidence interval bounds
    - sample_mean: Mean of the generated sample
    """
    # Generate data from a normal distribution
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Calculate sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation

    # Calculate standard error
    standard_error = sample_std / np.sqrt(size)

    # Find Z-score for the given confidence level
    z_score = norm.ppf((1 + confidence_level) / 2)

    # Calculate confidence interval
    margin_of_error = z_score * standard_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean


# Parameters for the normal distribution
mean = 50       # True mean
std_dev = 10     # True standard deviation
size = 100       # Number of data points
confidence_level = 0.95  # 95% confidence level

# Calculate confidence interval
lower_bound, upper_bound, sample_mean = confidence_interval_from_normal_data(
    mean, std_dev, size, confidence_level)

# Print results
print(f"Generated Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval ({confidence_level * 100:.0f}%): [{lower_bound:.2f}, {upper_bound:.2f}]")
```

```python
'''Q.23] Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.'''

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_cdf(lam, max_k):
    """
    Calculate and visualize the CDF of a Poisson distribution.

    Parameters:
    - lam: Rate parameter (λ) of the Poisson distribution
    - max_k: Maximum value of k for which to compute the CDF
    """
    # Define the range of k values (discrete values for Poisson)
    k = np.arange(0, max_k + 1)

    # Calculate the CDF
    cdf = poisson.cdf(k, mu=lam)

    # Plot the CDF
    plt.figure(figsize=(10, 6))
    plt.step(k, cdf, where='post', label=f'Poisson(λ={lam})', color='blue')
    plt.title("Cumulative Distribution Function (CDF) of Poisson Distribution")
    plt.xlabel("k (Number of events)")
    plt.ylabel("Cumulative Probability")
    plt.grid(alpha=0.3)
    plt.axhline(1.0, color='red', linestyle='--', label='P(X ≤ ∞) = 1')
    plt.legend()
    plt.show()

    # Interpretation
    print(f"Interpretation for Poisson(λ={lam}):")
    for ki, cdf_val in zip(k, cdf):
        print(f"P(X ≤ {ki}) = {cdf_val:.4f}")

# Parameters for the Poisson distribution
lam = 3        # Rate parameter (λ)
max_k = 10     # Maximum number of events

# Plot the CDF
plot_poisson_cdf(lam, max_k)
```

```python
'''Q.26] Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.'''
import numpy as np
import matplotlib.pyplot as plt

def calculate_range_and_iqr(data):
    """
    Calculate and display the range and interquartile range (IQR) of a dataset.
    
    Parameters:
    - data: Dataset (list or numpy array)
    """
    # Calculate range
    data_range = np.max(data) - np.min(data)
    
    # Calculate interquartile range (IQR)
    q1 = np.percentile(data, 25)  # 25th percentile (Q1)
    q3 = np.percentile(data, 75)  # 75th percentile (Q3)
    iqr = q3 - q1

    # Print results
    print(f"Range of the dataset: {data_range:.2f}")
    print(f"Interquartile Range (IQR) of the dataset: {iqr:.2f}")

    # Visualization
    plt.figure(figsize=(10, 6))
    plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor="skyblue", color="black"))
    plt.title("Box Plot of Dataset", fontsize=14)
    plt.xlabel("Values", fontsize=12)
    plt.grid(alpha=0.3)
    plt.show()

# Generate dataset from normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean = 50, Std Dev = 10, Size = 1000

# Calculate range and IQR
calculate_range_and_iqr(data)
```

```python
'''Q.27]  Implement Z-score normalization on a dataset and visualize its transformation.'''
import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    """
    Apply Z-score normalization to the dataset.

    Parameters:
    - data: Dataset (list or numpy array)
    
    Returns:
    - normalized_data: Z-score normalized dataset
    """
    mean = np.mean(data)  # Calculate mean
    std = np.std(data)  # Calculate standard deviation
    normalized_data = (data - mean) / std  # Apply Z-score normalization
    return normalized_data

# Generate dataset from normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean = 50, Std Dev = 10, Size = 1000

# Apply Z-score normalization
normalized_data = z_score_normalization(data)

# Visualization: Original vs. Normalized data
plt.figure(figsize=(14, 6))

# Histogram of Original Data
plt.subplot(1, 2, 1)
plt.hist(data, bins=20, color="skyblue", edgecolor="black", alpha=0.7)
plt.title("Original Data (Before Normalization)", fontsize=14)
plt.xlabel("Values", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(alpha=0.3)

# Histogram of Normalized Data
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=20, color="lightgreen", edgecolor="black", alpha=0.7)
plt.title("Z-Score Normalized Data", fontsize=14)
plt.xlabel("Z-Score", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(alpha=0.3)

# Display plots
plt.tight_layout()
plt.show()

# Print original and normalized stats
print(f"Original Mean: {np.mean(data):.2f}, Original Std Dev: {np.std(data):.2f}")
print(f"Normalized Mean: {np.mean(normalized_data):.2f}, Normalized Std Dev: {np.std(normalized_data):.2f}")
```
